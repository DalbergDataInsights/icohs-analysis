```python
import numpy as np
import pandas as pd
import geopandas as gpd
from pyproj import CRS
import difflib as dfb
```

```python
dis19_path = '../Data/Mapping layout/UBOS19 135 districts data/uganda_districts_2019_i.shp'
dis17_path = '../Data/Mapping layout/Humanitarian 120 districts data/Admin1/uga_admbnda_adm1_UBOS_v2.shp'
match_path = '../Data/Mapping layout/district files comparison.xlsx'
```

```python
dis19 = gpd.read_file(dis19_path)
dis19.to_crs(crs="EPSG:4326",inplace=True)
dis17 = gpd.read_file(dis17_path)
match= pd.read_excel(match_path,sheet_name=1)
```

```python
match.drop(['jenny_ 17_name','hum_name'],axis=1,inplace=True)
```

```python
district=dis19.merge(match,left_on='dc2017',right_on='jenny_ 17_code',how='left')
```

```python
district.drop('jenny_ 17_code',axis=1,inplace=True)
```

```python
name_dict = {'F15Regions': 'region',
'DName2016':'name_16',
'DNama2017':'name_17',
'dc2017':'code_17',
'DName2018':'name_18',
'dc2018':'code_18',
'DName2019':'name_19', 
'hum_code':'ADM1_PCODE'}
```

```python
district.rename(columns=name_dict,inplace=True)
```
